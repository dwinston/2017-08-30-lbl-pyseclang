```python
ch4 = pd.read_csv("data/atmo_CH4.csv", index_col="Year")
n2o = pd.read_csv("data/atmo_N2O.csv", index_col="Year")
```

```python
means = [chem.mean(axis="columns") for chem in [co2, ch4, n2o]]
```

```python
summary = pd.concat(means, axis="columns")
summary.columns = ["CO2", "CH4", "N2O"]
```

```python
summary = summary.fillna(method='bfill')
```

```python
summary.plot()
```

```python
## Add the temperature data to the summary dataframe
## Hint: use pd.concat, use Annual_Mean
summary = pd.concat([summary, temperature], axis = "columns")
```

```python
final = summary.dropna(how="any", axis="rows")
```

```python
final[["CO2", "Annual_Mean"]].plot(subplots=True)
```

```python
final.plot(subplots=True)
plt.show()
```

```python
final.corr()
```

```python
final.cov()
```

```python
import statsmodels.formula.api as sm
```

```python
model = sm.ols("Annual_Mean ~ CO2", data=final)
```

```python
fit = model.fit()
```

```python
fit.summary()
```

```python
from matplotlib import pyplot as plt
final.plot(y = "Annual_Mean", x = "CO2", kind="scatter")
plt.plot(final['CO2'], fit.predict(final), 'k-')
```

```python
final.to_csv("final.csv")
```

```python
sentence = "This is a sentence"
```

```python
sentence.split()
```

```python
lines = text.split('\n')
print(lines)
```

```python
lines = [line for line in lines if not line.startswith("#")]
```

```python
for line in lines:
    print(line.split())
```

```python
lines.pop(1)
```

```python
output_text = ''
for line in lines:
    output_line = line.split()
    output_line = ','.join(output_line) + '\n'
    print(output_line)
    output_text = output_text + output_line
```

```python
with open("output.csv", "w") as f:
    f.write(output_text)
```
